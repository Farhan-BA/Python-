```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Load dataset
df = pd.read_excel("/content/flight_price (1).xlsx")
```

```python
# First and last 5 rows
print(df.head())
print(df.tail())
```

```python
# Null values
print(df.isna().sum())
```

```python
# Info
print(df.info())
```

```python
# Describe
print(df.describe())
```

```python
# Drop unnecessary column
df.drop(['Route'], axis=1, inplace=True)
```

```python
# Fill missing values
df['Price'] = df['Price'].fillna(df['Price'].median())
df['Total_Stops'] = df['Total_Stops'].fillna(df['Total_Stops'].mode()[0])
df['Additional_Info'] = df['Additional_Info'].fillna('No info')
df['Duration'] = df['Duration'].fillna('0h 0m')
```

```python
# Convert Duration to minutes
def convert_duration(x):
    try:
        x = x.strip()
        h, m = 0, 0
        if 'h' in x:
            h = int(x.split('h')[0])
            if 'm' in x:
                m = int(x.split('h')[1].replace('m', '').strip())
        elif 'm' in x:
            m = int(x.replace('m', '').strip())
        return h * 60 + m
    except:
        return np.nan

df['Duration_Minutes'] = df['Duration'].apply(convert_duration)
```

```python
# Fix Total_Stops
df['Total_Stops'] = df['Total_Stops'].replace({
    'non-stop': 0,
    '1 stop': 1,
    '2 stops': 2,
    '3 stops': 3,
    '4 stops': 4,
    'NaN': 0
})
df['Total_Stops'] = pd.to_numeric(df['Total_Stops'], errors='coerce').fillna(0)
```

```python
# Convert datetime columns
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'], errors='coerce')
df['Dep_Time'] = pd.to_datetime(df['Dep_Time'], errors='coerce')
df['Arrival_Time'] = pd.to_datetime(df['Arrival_Time'], errors='coerce')
```

```python
# Extract features
df['Journey_Day'] = df['Date_of_Journey'].dt.day
df['Journey_Month'] = df['Date_of_Journey'].dt.month
df['Dep_Hour'] = df['Dep_Time'].dt.hour
df['Arrival_Hour'] = df['Arrival_Time'].dt.hour
```

```python
# Feature engineering
df['Time_of_Day'] = df['Dep_Hour'].apply(
    lambda x: 'Morning' if 5 <= x < 12 else 'Afternoon' if 12 <= x < 17 else 'Evening' if 17 <= x < 21 else 'Night'
)
df['Is_Weekend'] = df['Date_of_Journey'].dt.dayofweek >= 5
df['Is_Rush_Hour'] = df['Dep_Hour'].apply(lambda x: 1 if x in [7, 8, 17, 18, 19] else 0)
```

```python

# Encode categorical columns
df = pd.get_dummies(df, columns=['Airline', 'Source', 'Destination', 'Additional_Info', 'Time_of_Day'], drop_first=True)

```

```python
# Drop unused original columns
df.drop(['Date_of_Journey', 'Dep_Time', 'Arrival_Time', 'Duration'], axis=1, inplace=True)
```

```python

# Ensure all columns are numeric
non_numeric_cols = df.select_dtypes(exclude=['number']).columns
if len(non_numeric_cols) > 0:
    print("Warning: Non-numeric columns found:", non_numeric_cols)
```

```python
# Drop any remaining NA
df.dropna(inplace=True)
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Duration_Minutes'], kde=True)
plt.title("Flight Duration Distribution")
plt.show()
```

```python
# Price vs Duration
sns.scatterplot(x='Duration_Minutes', y='Price', data=df)
plt.title("Price vs Duration")
plt.show()
```

```python
# Select only numeric features of interest
corr_features = ['Price', 'Total_Stops', 'Duration_Minutes', 'Journey_Day', 'Journey_Month',
                 'Dep_Hour', 'Arrival_Hour']
```

```python
# Optional: include engineered features if you added them
if 'Is_Weekend' in df.columns and 'Is_Rush_Hour' in df.columns:
    corr_features += ['Is_Weekend', 'Is_Rush_Hour']
```

```python
# Compute correlation
corr_matrix = df[corr_features].corr()
```

```python
# Plot
plt.figure(figsize=(10, 7))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap - Numeric Features Only", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Time of Day
df['Time_of_Day'] = df['Dep_Hour'].apply(lambda x: 'Morning' if 5<=x<12
                                         else 'Afternoon' if 12<=x<17
                                         else 'Evening' if 17<=x<21
                                         else 'Night')
```

```python
# Weekday vs Weekend
df['Day_of_Week'] = pd.to_datetime(df['Journey_Day'], errors='coerce').dt.dayofweek
df['Weekend'] = df['Day_of_Week'].apply(lambda x: 1 if x>=5 else 0)
```

```python

# Rush Hour (Morning or Evening Flights)
df['Is_Rush_Hour'] = df['Time_of_Day'].apply(lambda x: 1 if x in ['Morning', 'Evening'] else 0)
```
